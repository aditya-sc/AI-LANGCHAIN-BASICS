### integrate models

```python
# init_chat_model is the provider-agnostic way to load a model: pass
# "provider:model" ("ollama:qwen3:8b") and it returns a ready chat model. The next
# cell uses the provider-specific ChatOllama directly — interchangeable here.
from langchain.chat_models import init_chat_model
model=init_chat_model("ollama:qwen3:8b")
model
```

```python
## invoke = blocking, returns the whole answer at once. Fine for dev/notebooks;
## in production prefer streaming so output feels instant (NOTES.md §11).
response=model.invoke("How are you ?")
response
```

```python
from langchain_ollama import ChatOllama
model=ChatOllama(model="qwen3:8b")
response=model.invoke("How are you ?")
response
```

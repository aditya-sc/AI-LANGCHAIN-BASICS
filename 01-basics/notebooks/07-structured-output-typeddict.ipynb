### typeddict

```python
# --- Structured output with TypedDict (NOTES.md §9) ---
# TypedDict is the lightweight alternative to Pydantic: it's just a dict + type hints,
# no runtime validation. Here Annotated[type, ..., "description"] supplies the same
# field descriptions the model needs. Good for graph state; for untrusted LLM output
# Pydantic (with its validation/coercion) is usually the safer choice.
from typing_extensions import TypedDict, Annotated
from langchain.chat_models import init_chat_model

class MovieDict(TypedDict):
    """A movie with details"""
    title: Annotated[str, ..., "The title of the movie"]
    year: Annotated[int, ..., "The year movie was released"]
    director: Annotated[str, ..., "The director of the movie"]
    rating: Annotated[float, ..., "The rating of the movie out of 10"]

model=init_chat_model("ollama:qwen3:8b")
model.with_structured_output(MovieDict).invoke("Tell about movie Inception")
```

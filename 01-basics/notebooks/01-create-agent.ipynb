```python
# --- create_agent: the standard ReAct loop in ONE call (NOTES.md §6) ---
from langchain.agents import create_agent
from langchain_ollama import ChatOllama

def get_weather(city:str) -> str:
    """get weather for a city."""   # docstring = the description the LLM uses to decide calls
    return f"the weather in {city} is sunny."

llm = ChatOllama(
    model="qwen3:8b"
)

# create_agent (LangChain 1.0) is a factory that builds the whole StateGraph for you
# — add_node/tools_condition/compile wiring included — and returns the same kind of
# compiled graph you'd hand-build. It gives ONE fixed topology (model + tools, loop
# until done); you configure it (model/tools/prompt) rather than rearranging nodes.
# Hand-build a StateGraph only when you need ordering/branching/parallel/approval/retry
# guarantees (NOTES.md §6).
agent = create_agent(
    model=llm,
    tools=[get_weather],
    system_prompt="You are a helpful assistant."
)
agent
```

```python
### run the agent
response = agent.invoke({"messages":[{"role":"user","content":"what is weather like in mumbai?"}]})
```
